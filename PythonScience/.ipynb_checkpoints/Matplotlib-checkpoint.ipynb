# Matplotlib Learning Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np
x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
'''it simply takes one command to draw.. '''
pylab.plot(x, y, 'r') # r stands for red
```

```python
'''Drawing Sub graph '''
pylab.subplot(1, 2, 1) # The content of bracket represent (rows, columns, indexes)
pylab.plot(x, y, 'r--')# the third parameter here determines colour and line style
pylab.subplot(1, 2, 2)
pylab.plot(y, x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
# again we can draw subgraphs
fig, axes = plt.subplots(nrows = 1, ncols = 3) #Submap is 1 row, 2 columns 
for ax in axes:
    ax.plot(x, y, 'r')
```

```python
# we can also draw, or graph, inside another graph
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small convas 
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
```

```python
fig = plt.figure(figsize=(16,9), dpi = 50) # new graphic object dpi = picture qulity
fig.add_subplot()
plt.plot(x, y, 'r')
```

```python
ax.legend(['label1, label2'])

fig, axes = plt.subplots()

axes.set_title('Title of the graph')
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(['y = x**2', 'y = x**3'], loc = 2)  # loc defines position of legend remark on the graph plot
```

```python
fig , axes = plt.subplots(dpi = 100)
axes.plot(x, x**2, color = 'green', alpha=.5)
axes.plot(x, x+2, color = 'blue', alpha=.5)
axes.plot(x, x+3, color = 'orange', alpha=.5)
```

```python
fig , axes = plt.subplots(dpi = 100)
axes.plot(x, x+1, color = 'blue', linewidth = 0.25)
axes.plot(x, x+2, color = 'blue', linewidth = 1)
axes.plot(x, x+3, color = 'blue', linewidth = 1.5)
axes.plot(x, x+4, color = 'blue', linewidth = 2)
```

```python
fig , axes = plt.subplots(dpi = 100)
axes.plot(x, x+1, color = 'blue', lw=2, linestyle = '-')
axes.plot(x, x+2, color = 'blue', lw=2, linestyle = '-.')
axes.plot(x, x+3, color = 'blue', lw=2, linestyle = ':')
# '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

line, = axes.plot(x, x+4, color = 'black', lw = 1.50)
line.set_dashes([5, 10, 15, 10])

line, = axes.plot(x, x+5, color = 'black', lw = 1.50)
line.set_dashes([5, 6, 12, 10])

line, = axes.plot(x, x+6, color = 'black', lw = 1.50)
line.set_dashes([5, 8, 15, 12])

line, = axes.plot(x, x+7, color = 'black', lw = 1.50)
line.set_dashes([6, 10, 11, 10])
```

```python
fig , axes = plt.subplots(dpi = 80)
axes.plot(x, x+1, color = 'blue', marker = '+', markersize = 5, markerfacecolor = 'red')
axes.plot(x, x+2, color = 'blue', marker = 'x', markersize = 5, markerfacecolor = 'red')
axes.plot(x, x+3, color = 'blue', marker = '*', markersize = 5, markerfacecolor = 'red')
```

```python
# set the convas grid and axis range

fig, Axes = plt.subplots(1, 2, figsize = (10, 5))  # 1, 2 argument in subplots declare 2 plots named 1 and 2

Axes[0].plot(x, x**2, x, x**3, lw = 2)
Axes[0].grid(True)

Axes[1].plot(x, x**2, x, x**3, lw = 2)
Axes[1].set_ylim([0, 60])
Axes[1].set_xlim([2, 5])
Axes[1].grid(True)
```

# Other 2D graphics

```python
n = np.arange(6)

fig, axes = plt.subplots(1, 4, figsize = (16, 5))

axes[0].set_title('scatter')
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title('Step')
axes[1].step(n, n**2, lw = 2)

axes[2].set_title('Bar')
axes[2].bar(n, n**2, align = 'center', width = 0.5, alpha = 0.8)

axes[3].set_title('Fill_Between')
axes[3].fill_between(x, x**2, x**3,  color = 'green', alpha = 0.8)
```

```python
# Draw a radar chart

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, .6, .6], polar = True)

t = np.linspace(0, 2*np.pi, 100)

ax.plot(t, t, color = 'blue', lw = 3)
```

```python
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].set_title('Default Histogram')
axes[0].hist(n, bins = 8)

axes[1].set_title('Cumulative detailed Histogram')
axes[1].hist(n, cumulative = True, bins = 50)
```
